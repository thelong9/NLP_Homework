# Compute term-document matrix

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("data/all_wiki_datas.csv")
```

```python
def tokenizer(data: str):
    data = data.lower()
    data = re.sub(r"[\W_]", " ", data)
    tokens = data.split()
    return tokens
```

```python
def create_vocab_and_document_tokens(df: pd.DataFrame):
    vocaburary = set()
    document_tokens = []
    
    for idx, row in df.iterrows():
        text = row['text']
        tokens = tokenizer(text)
        document_tokens.append(tokenizer(text))
        vocaburary.update(tokenizer(text))
    return sorted(list(vocaburary)), document_tokens
```

```python
vocab, document_tokens = create_vocab_and_document_tokens(df)
```

```python
columns = df['title'].to_list()
rows = vocab

term_doc_df = pd.DataFrame(0, columns=columns, index= rows)
```

```python
for idx, row in df.iterrows():
    tokens_per_doc = tokenizer(row['text'])
    for token in tokens_per_doc:
        term_doc_df.at[token, row['title']] += 1
    
```
