Create a dataset from 50 Wikipedia articles. The articles should cover several topics, such as
Science, Politics, Nature, Sport, and Fashion.
Copy the text in each article into a text file with the name in the following format: ID - Article
title.txt. For example, 1 - Donald Trump.txt.
Put the files in a folder named Data
Zip the folder to Data.zip


```python
import pandas as pd
```

```python
from datasets import load_dataset

ds = load_dataset("wikimedia/wikipedia", "20231101.en", streaming=True)
```

```python
ds = ds['train']
```

```python
ds = ds.take(5000)
```

```python
df = pd.DataFrame.from_dict(ds)
```

```python
df = df.sample(50)
```

```python
df
```

```python
df.to_csv('all_wiki_datas.csv', index=False)
```

```python
for idx, row in df.iterrows():
    id = row['id']
    title = row['title']
    text = row['text']
    
    with open(f"data/{id} - {title}.txt", "w") as f:
        f.write(text)
    
    
```

```python
import zipfile
import os

folder_to_zip = "data"

zipfile_name = "data.zip"

with zipfile.ZipFile(zipfile_name, 'w') as data_zip:
    for foldername, subfolders, filenames in os.walk(folder_to_zip):
        for filename in filenames:
            file_path = os.path.join(foldername, filename)
            
            data_zip.write(file_path, os.path.relpath(file_path,folder_to_zip))
```
