# Compute the tf-idf for each term in the vocabulary

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("result/term_document_matrix.csv")
```

```python
df = df.rename(columns={'Unnamed: 0' : 'Token'})
```

```python
df = df.set_index('Token')
```

```python
total_doc = 50
```

```python
def cal_tf(frequency: pd.Series) -> np.array:
    arr = frequency.to_numpy()
    arr = np.log10(1 + arr)
    return arr 
def cal_df(frequency: pd.Series) -> np.array:
    n = np.sum(frequency > 0, axis= 0)
    return n
def cal_idf(frequency: pd.Series) -> np.array:
    n = frequency.shape[0]
    df = cal_df(frequency)
    idf = np.log10(n/df)
    return idf
def cal_tf_idf(frequency: pd.Series) -> np.array:
    tf = cal_tf(frequency)
    idf = cal_idf(frequency)
    tf_idf = tf * idf
    return tf_idf
    
```

```python
df
```

```python
tf_idf_df = df.apply(cal_tf_idf, axis = 0)
```
